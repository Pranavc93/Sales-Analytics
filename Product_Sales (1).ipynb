```python
all_months_sales.columns = all_months_sales.columns.str.replace('Order Date','OrderDate')
all_months_sales.columns = all_months_sales.columns.str.replace('Purchase Address','PurchaseAddress')
```

```python
#Create new columns by splitting the order date to get Year, Month and Date columns

all_months_sales['Year'] = all_months_sales['OrderDate'].dt.year
```

```python
all_months_sales['Month'] = all_months_sales['OrderDate'].dt.month
```

```python
all_months_sales['Date'] = all_months_sales['OrderDate'].dt.day
```

```python
all_months_sales['Hour'] = all_months_sales['OrderDate'].dt.hour
```

```python
all_months_sales['Minute'] = all_months_sales['OrderDate'].dt.minute
```

```python
all_months_sales
```

```python
#Now split the purchase address column  into different columns like house number, street, city, state, postal code

all_months_sales [['HouseStreet','City','State']] = all_months_sales.PurchaseAddress.str.split(",", expand = True)
```
